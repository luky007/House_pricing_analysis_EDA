```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import os
from pathlib import Path
import seaborn as sns
import matplotlib.pyplot as plt
from pandas_dq import dq_report

#set dark mode
%matplotlib inline
sns.set(style="ticks", context="talk")
plt.style.use("dark_background")

pd.options.display.float_format = "{:20,.2f}".format
current_project_dir = Path(os.path.abspath(""))
path_csv_file = Path(os.path.abspath("")).parent.parent / "output" / "data_house_merged.csv"
path_report_html = Path(os.path.abspath("")).parent.parent / "output" / "report.html"
path_report_clean_html = Path(os.path.abspath("")).parent.parent / "output" / "report_clean.html"

data = pd.read_csv(path_csv_file)

#convert furnishingstatus
furnishing_mapping = {
    "unfurnished": 0,
    "semi-furnished": 1,
    "furnished": 2
}
data["furnishing_status_numeric"] = data["furnishingstatus"].map(furnishing_mapping)
data = data.drop(["ID","furnishingstatus"], axis=1)

# convert yes and no to bool
bool_columns = ["mainroad", "guestroom", "basement", "hotwaterheating", "airconditioning", "prefarea"]
data[bool_columns] = data[bool_columns].apply(lambda x: x.map({"yes": 1, "no": 0}))

# convert int
#data["price"] = data["price"] // 100
#data["area"] = data["area"] / 100
data["price"] = data["price"] * 0.0111792 # indian rupies to euro
data["area"] = data["area"] * 0.092903  # conversion factor between square feet and square meters

print(data.head())
print(data.tail())
print(f"Total null values in the DataFrame: {data.isnull().sum().sum()}")
```

```python
def remove_outliers(df, column_name, multiplier=1.5):
    # avoid removing from hotwaterheating and mainroad
    if column_name in ["mainroad","hotwaterheating"]:
        return df

    # Calculate the IQR for the specified column
    quartiles = df[column_name].quantile([0.25, 0.75])
    iqr = quartiles[0.75] - quartiles[0.25]

    # Calculate the lower and upper bounds
    lower_bound = quartiles[0.25] - multiplier * iqr
    upper_bound = quartiles[0.75] + multiplier * iqr

    # Create a boolean mask for outliers
    outliers_mask = (df[column_name] < lower_bound) | (df[column_name] > upper_bound)

    # Filter the DataFrame to exclude outliers
    filtered_df = df[~outliers_mask]
    return filtered_df

filtered_data = data.copy()
while True:
    element_before_cleaning = len(filtered_data)
    # Remove outliers for all numeric columns
    for col in data.select_dtypes(include="number"):
        filtered_data = remove_outliers(filtered_data, col)
    if len(filtered_data) == element_before_cleaning:
        break
# avoid removing from hotwaterheating and mainroad
filtered_data["hotwaterheating"] = data["hotwaterheating"]
filtered_data["mainroad"] = data["mainroad"]
```

```python
dqr = dq_report(data, target=None, html=False, csv_engine="pandas", verbose=1)
dqr = dq_report(filtered_data, target=None, html=False, csv_engine="pandas", verbose=1)
```

```python
# Group by "hotwaterheating" and count the occurrences of "yes" and "no"
hotwater_counts = data["hotwaterheating"].map({
    0: 'no',
    1: 'yes'
}).value_counts()

# Group by "mainroad" and count the occurrences of "yes" and "no"
mainroad_counts = data["mainroad"].map({
    0: 'no',
    1: 'yes'
}).value_counts()

# Create a bar chart
fig, ax = plt.subplots(figsize=(8, 6))
hotwater_counts.plot(kind="bar", ax=ax, color="skyblue", label="Hot Water Heating",alpha=0.5)
mainroad_counts.plot(kind="bar", ax=ax, color="salmon", label="Main Road",alpha=0.5, width=0.2)
ax.set_xlabel("Feature")
ax.set_ylabel("Count")
ax.set_title("Distribution of Hot Water Heating and Main Road Features")
plt.legend()
plt.show()
print(f"Percentage of properties with hot water heating: {(hotwater_counts['yes'] / len(data)) * 100:.2f}")
print(f"Percentage of properties with no hot water heating: {(hotwater_counts['no'] / len(data)) * 100:.2f}")
print(f"Percentage of properties located near the main road: {(mainroad_counts['yes'] / len(data)) * 100:.2f}")
print(f"Percentage of properties located not near the main road: {(mainroad_counts['no'] / len(data)) * 100:.2f}")
```

```python
x = filtered_data[["area"]]
y = filtered_data["price"]

# Initialize and train the linear regression model
model = LinearRegression()
model.fit(x, y)

# Make predictions
y_pred = model.predict(x)

# Calculate R2 score
r2 = r2_score(y, y_pred)
print(f"R2 Score: {r2:.4f}")
```

```python
# Calculate the average presence of basements for each number of stories
avg_basement_by_stories = filtered_data.groupby("stories")["basement"].mean()

# Normalize the average basement presence to [0, 1] range
normalized_avg_basement = (avg_basement_by_stories - avg_basement_by_stories.min()) / (avg_basement_by_stories.max() - avg_basement_by_stories.min())

# Scale down the number of houses
scaled_num_houses = filtered_data.groupby("stories").size() / filtered_data.shape[0]

# Create a grouped bar plot
plt.figure(figsize=(10, 6))
plt.bar(avg_basement_by_stories.index, normalized_avg_basement, label="Avg Basement Presence", alpha=0.4)
plt.bar(avg_basement_by_stories.index, scaled_num_houses, label="Number of Houses", alpha=0.5, width=0.4)

plt.title("Basement Presence and Number of Houses by Number of Stories")
plt.xlabel("Number of Stories")
plt.ylabel("Normalized Value")
plt.legend()

plt.show()
```
